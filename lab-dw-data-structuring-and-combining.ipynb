In this challenge, we will be working with the customer data from an insurance company, as we did in the two previous labs. The data can be found here:

Note that you'll need to clean and format the new data.

Observation:

- Another option would be to read the clean file you saved in the previous lab, and just clean the two new files and concatenate the three clean datasets

```python
# Your code goes here
import pandas as pd
data1 = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv")
data2 = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv")
data3= pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv")
```

```python
data1.rename(columns={'ST': 'State', 'GENDER': 'Gender'}, inplace=True)
data1.head()
```

```python
data2.rename(columns={'ST': 'State', 'GENDER': 'Gender'}, inplace=True)
data2.head()
```

```python
data3.head()
```

```python
new_data = pd.concat([data1, data2, data3], axis=0)
new_data
```

```python
new_data.drop_duplicates(inplace = True)
```

```python
new_data.rename(columns={'Customer': 'customer', 'State': 'state', 'Gender':'gender','Education':'education','Customer Lifetime Value':'customer_lifetime_value','Income':'income',
                      'Monthly Premium Auto':'monthly_premium_auto','Number of Open Complaints':'number_of_open_complaints','Policy Typer':'policy_type',
                     'Vehicle Class':'vehicle_class','Total Claim Amount':'total_claim_amount'}, inplace=True)
```

```python
new_data["gender"] = new_data["gender"].apply(lambda x :  "F" if x == "female" else x)
new_data["gender"] = new_data["gender"].apply(lambda x :  "F" if x == "Femal" else x)
new_data["gender"] = new_data["gender"].apply(lambda x :  "M" if x == "Male" else x)
new_data["gender"].unique()
```

```python
new_data["state"] = new_data["state"].apply(lambda x :  "Washington" if x == "WA" else x)
new_data["state"] = new_data["state"].apply(lambda x :  "Arizona" if x == "AZ" else x)
new_data["state"] = new_data["state"].apply(lambda x :  "California" if x == "Cali" else x)
new_data["state"].unique()
```

```python
new_data["customer_lifetime_value"] = new_data["customer_lifetime_value"].apply(lambda x :  x[:-1] if (isinstance(x,str) and x[-1] == "%") else x)
new_data["customer_lifetime_value"] = new_data["customer_lifetime_value"].apply(lambda x :  float(x) if pd.isna(x) == False else x)
```

```python
new_data["vehicle_class"] = new_data["vehicle_class"].apply(lambda x :  "Luxury" if x == "Sports Car" else x)
new_data["vehicle_class"] = new_data["vehicle_class"].apply(lambda x :  "Luxury" if x == "Luxury SUV" else x)
new_data["vehicle_class"] = new_data["vehicle_class"].apply(lambda x :  "Luxury" if x == "Luxury Car" else x)
new_data["vehicle_class"].unique()
```

```python
new_data["number_of_open_complaints"] = new_data["number_of_open_complaints"].apply(lambda x: int(x.split("/")[1]) if isinstance(x, str) and len(x.split("/")) == 3 else x)
new_data.isnull().sum()
```

```python
new_data = new_data.dropna()
new_data.isnull().sum()
```

```python
new_data["number_of_open_complaints"] = new_data["number_of_open_complaints"].apply(lambda x : int(x))
new_data
```

1. You work at the marketing department and you want to know which sales channel brought the most sales in terms of total revenue. Using pivot, create a summary table showing the total revenue for each sales channel (branch, call center, web, and mail).
Round the total revenue to 2 decimal points.  Analyze the resulting table to draw insights.

```python
structuring_data_pivot = structuring_data.pivot_table( columns="sales_channel", values="customer_lifetime_value", aggfunc = "sum")
structuring_data_pivot = structuring_data_pivot.applymap(lambda x: f"{x:,.2f}")
structuring_data_pivot
```

```python
structuring_data_pivot2 = structuring_data.pivot_table(index = "gender", columns="education", values="customer_lifetime_value", aggfunc = "mean")
structuring_data_pivot2
```
